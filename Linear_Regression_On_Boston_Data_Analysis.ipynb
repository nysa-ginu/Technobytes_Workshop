## Data Split into Train and Test
#### This is also known as Cross Validation

```python
X=boston_df.drop(['MEDV'],axis=1)
Y=boston_df['MEDV']
from sklearn.model_selection import train_test_split 
# will train and test X and Y separately
train_x , test_x , train_y , test_y=train_test_split(X,Y,test_size=0.15,random_state=2) # for random_state = 0 lr_model.score = 0.54 ie 54% accuracy and for random_state=2 accuracy = 75% which is excellent for regression.
#random_state is to initiate randomization to avoid overfitting
```

```python
train_x.shape
```

```python
train_y.shape
```

```python
from sklearn.linear_model import LinearRegression
lr_model=LinearRegression()
lr_model.fit(train_x,train_y)
```

```python
yhat=lr_model.predict(test_x)
```

```python
lr_model.score(test_x,test_y)
```

```python
plt.figure(figsize=(12,8))
plt.plot(yhat)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score   #these two are the statistocs which gives us the final outcome
print(mean_squared_error(test_y,yhat))   #the lower the mean squared error higher the accuracy
print(r2_score(test_y,yhat))  #yhat is the predicted value of y
```
